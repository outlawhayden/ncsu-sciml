```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = np.load("/Users/haydenoutlaw/Documents/Courses/SciML/ncsu-sciml/hw3/data/learning_trajectories_20251015_102417.npz")
ds.keys
```

```python
u_2000 = ds['u_2000_log_loss']
u_4000 = ds['u_4000_log_loss']
u_6000 = ds['u_6000_log_loss']
m_2000 = ds['m_2000_log_loss']
m_4000 = ds['m_4000_log_loss']
m_6000 = ds['m_6000_log_loss']
l_2000 = ds['l_2000_log_loss']
l_4000 = ds['l_4000_log_loss']
l_6000 = ds['l_6000_log_loss']
```

```python
num_epochs = np.arange(1, m_2000.shape[0] + 1)
sns.set_theme(rc={'figure.figsize':(15,5)}, font_scale = 1.3, context = 'notebook')



sns.lineplot(x= num_epochs, y = u_2000, label = 'uniform, 2000')
sns.lineplot(x= num_epochs, y = u_4000, label = 'uniform, 4000')
sns.lineplot(x= num_epochs, y = u_6000, label = 'uniform, 6000')
sns.lineplot(x= num_epochs, y = m_2000, label = 'mesh, 2000')
sns.lineplot(x= num_epochs, y = m_4000, label = 'mesh, 4000')
sns.lineplot(x= num_epochs, y = m_6000, label = 'mesh, 6000')
sns.lineplot(x= num_epochs, y = m_2000, label = 'LHS, 2000')
sns.lineplot(x= num_epochs, y = m_4000, label = 'LHS, 4000')
sns.lineplot(x= num_epochs, y = m_6000, label = 'LHS, 6000')


plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.yscale("log")
plt.title("Loss across Epochs, L2 Loss on Test Dataset")
plt.legend()
plt.tight_layout()
plt.savefig("grid_experiment.png", dpi = 300)
plt.show()
```
